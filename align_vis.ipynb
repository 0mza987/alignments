```python
FILE_PATH = '/home/ubuntu/Downloads/180727-off-line-english-init-model.save_epoch2.log'
LIST_lines = open(FILE_PATH).readlines()
LIST_ERR = [float(line.split(', file:')[0].split('err:')[1].strip()) for line in LIST_lines]
print 'Total lines: {}'.format(len(LIST_ERR))
LIST_err = [err for err in LIST_ERR if err>=0.5]
print 'Above 0.5: {}'.format(len(LIST_err))
se = pd.Series(LIST_err)
```

```python
fig, ax = plt.subplots(figsize=(15,15))
g = sns.kdeplot(se, color="Blue", label='distribution',shade= True)
```
